# Análise de Protocolos e Portas

```python
# Extrair mais informações do campo Info
import re

def extract_tcp_flags(info):
    flags = []
    if 'SYN' in str(info): flags.append('SYN')
    if 'ACK' in str(info): flags.append('ACK')
    if 'FIN' in str(info): flags.append('FIN')
    if 'PSH' in str(info): flags.append('PSH')
    if 'RST' in str(info): flags.append('RST')
    return ','.join(flags) if flags else 'NONE'

def extract_seq_win(info):
    seq_match = re.search(r'Seq=(\d+)', str(info))
    win_match = re.search(r'Win=(\d+)', str(info))
    seq = int(seq_match.group(1)) if seq_match else 0
    win = int(win_match.group(1)) if win_match else 0
    return seq, win

# Aplicar funções
df['TCP_Flags'] = df['Info'].apply(extract_tcp_flags)
df[['Seq_Num', 'Win_Size']] = df['Info'].apply(
    lambda x: pd.Series(extract_seq_win(x))
)

# Criar features de fluxo
df['Flow_ID'] = df['Source'] + '_' + df['Destination']
df['Packet_Direction'] = df.apply(
    lambda x: 'OUT' if x['Source'] == '172.17.0.1' else 'IN',
    axis=1
)

# Features temporais avançadas
df['Time_Diff'] = df['Time'].diff()
df['Interarrival_Time'] = df.groupby('Flow_ID')['Time'].diff()
```

 Classificação de Tipo de Tráfego
# Objetivo: Classificar se o tráfego é normal ou suspeito/anômalo

```python
import pandas as pd # Import pandas
from sklearn.preprocessing import StandardScaler, LabelEncoder # Import LabelEncoder
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier # RandomForestClassifier was used but not imported
import re # Needed for port extraction

# --- Ensure df is loaded (for robustness) ---
df = pd.read_csv('/content/dockertorvpn.csv')

# --- Re-create Source_Port, Dest_Port, and Time_Diff columns (as they are used here and were removed from previous cells) ---
def extract_ports(info):
    ports = re.findall(r'(\d+) > (\d+)', str(info))
    if ports:
        return int(ports[0][0]), int(ports[0][1])
    return None, None

df[['Source_Port', 'Dest_Port']] = df['Info'].apply(
    lambda x: pd.Series(extract_ports(x))
)

# Convert to numeric, coercing errors (e.g., None -> NaN)
df['Source_Port'] = pd.to_numeric(df['Source_Port'], errors='coerce')
df['Dest_Port'] = pd.to_numeric(df['Dest_Port'], errors='coerce')

df['Time_Diff'] = df['Time'].diff().fillna(0) # Also ensure Time_Diff is present and handle NaNs

# Criar labels baseadas nas portas (aprox. serviços)
def label_service(row):
    dest_port = row['Dest_Port']
    if pd.isna(dest_port):
        return 'unknown'

    dest_port = int(dest_port)
    if dest_port == 443 or dest_port == 8443:
        return 'https'
    elif dest_port == 80:
        return 'http'
    elif dest_port == 53:
        return 'dns'
    elif dest_port == 9050 or dest_port == 9051:  # TOR
        return 'tor'
    elif 1024 <= dest_port <= 49151:
        return 'registered'
    else:
        return 'ephemeral'

# Criar dataset para classificação
df['service_type'] = df.apply(label_service, axis=1)

# --- Diagnostic prints ---
print("\nDistribution of service_type:")
print(df['service_type'].value_counts())

# Features para classificação
classification_features = pd.DataFrame({
    'packet_size': df['Length'],
    'protocol_encoded': LabelEncoder().fit_transform(df['Protocol']),
    'time_diff': df['Time_Diff'].fillna(0),
    'src_port': df['Source_Port'].fillna(0),
    'dst_port': df['Dest_Port'].fillna(0),
    'has_syn': df['Info'].str.contains('SYN').astype(int),
    'has_ack': df['Info'].str.contains('ACK').astype(int),
    'has_fin': df['Info'].str.contains('FIN').astype(int)
})

# Remover linhas com NaN
valid_idx = classification_features.notna().all(axis=1)
X = classification_features[valid_idx]
y = df['service_type'][valid_idx]

# Codificar labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# --- Diagnostic print for y_encoded ---
print("\nUnique values in y_encoded:", np.unique(y_encoded))

# Testar diferentes classificadores
classifiers = {
    'XGBoost': XGBClassifier(random_state=42, n_estimators=100),
    'LightGBM': LGBMClassifier(random_state=42, n_estimators=100),
    'RandomForest': RandomForestClassifier(random_state=42, n_estimators=100)
}

for name, clf in classifiers.items():
    # Only run classification if there's more than one class to predict
    if len(np.unique(y_encoded)) > 1:
        scores = cross_val_score(clf, X, y_encoded, cv=5, scoring='f1_weighted')
        print(f"{name}: F1 médio = {scores.mean():.3f} (+/- {scores.std():.3f})")
    else:
        print(f"{name}: Cannot perform cross-validation: Target variable has only one class after encoding.")
```

# Detecção de Anomalias

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt # Import plt for plotting

# Preparar dados
# Use 'classification_features' as it contains 'packet_size' and 'time_diff' and was previously defined.
X = classification_features[['packet_size', 'time_diff']].fillna(0)
X_scaled = StandardScaler().fit_transform(X)

# Treinar modelo de detecção de anomalias
model = IsolationForest(contamination=0.1, random_state=42)
df['anomaly_score'] = model.fit_predict(X_scaled)

# Visualizar anomalias
plt.figure(figsize=(12,6))
plt.scatter(df.index, df['Length'],
            c=df['anomaly_score'],
            cmap='coolwarm', alpha=0.6)
plt.title('Detecção de Anomalias no Tamanho dos Pacotes')
plt.colorbar(label='Anomalia (1=normal, -1=anômalo)')
```

# Previsão de Volume de Trafégo

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.linear_model import LinearRegression

# Criar série temporal
df['time_window'] = (df['Time'] // 1).astype(int)  # janelas de 1 segundo
traffic_volume = df.groupby('time_window').agg({
    'Length': 'sum',
    'No.': 'count'
}).reset_index()

# Preparar dados para previsão
X = traffic_volume['time_window'].values.reshape(-1, 1)
y = traffic_volume['Length'].values

# Dividir dados temporais
tscv = TimeSeriesSplit(n_splits=5)
for train_idx, test_idx in tscv.split(X):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    model = LinearRegression()
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)

    # Avaliar
    print(f"MSE: {mean_squared_error(y_test, predictions)}")
    print(f"R²: {r2_score(y_test, predictions)}")
```

# Clusterização de padroẽs de Comportamento

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Criar features agregadas por IP
ip_features = df.groupby('Source').agg({
    'Length': ['mean', 'std', 'sum'],
    'Time': ['count', 'mean'],
    'Destination': 'nunique'
}).fillna(0)

ip_features.columns = ['_'.join(col).strip() for col in ip_features.columns.values]

# Redução de dimensionalidade e clusterização
X_scaled = StandardScaler().fit_transform(ip_features)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Visualizar clusters
plt.figure(figsize=(10,8))
plt.scatter(X_pca[:,0], X_pca[:,1], c=clusters, cmap='viridis')
plt.title('Clusterização de Comportamento dos IPs de Origem')
```
